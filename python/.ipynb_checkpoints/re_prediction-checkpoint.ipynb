```python
with open('LOGLOGLOG.pickle', 'rb') as fr:
    LOG = pickle.load(fr)

training_features = ['ls1','g1', 'g2','p1','p2','year']
```

```python
with open('clog0.pickle', 'rb') as fr:
    LOG = pickle.load(fr)
model=LOG
cc=c0
sl=pd.read_csv("sleuthOutput.csv").iloc[:,1:]
for i in range(5):
    dataTest=datas[4-i]
    y50=pd.DataFrame(dataTest.iloc[cc,[0,1,3,4,6,24,5,15]])
    y50=y50.reset_index().iloc[:,1:]
    n50=pd.DataFrame({"year":np.ones(y50.shape[0])*50})
    y50=pd.concat([y50,n50],axis=1,ignore_index=True)
    y50.columns=['city', 'country','Regional_label', 'ls1', "p1",'p2', "g1",'g2', 'year']
    data2=y50
    data2.iloc[:,3:8]=np.log(data2.iloc[:,3:8])
    pred2050=np.exp(model.predict(data2[training_features])).astype('int')
    pred20502=pd.DataFrame({'city':data2.iloc[:,0],'ssp'+str(5-i):pred2050})

    sl=pd.merge(pred20502,sl,on='city')
    
sl.to_csv("sl_LOG_c0.csv")
# sl
sl.sum()
```
