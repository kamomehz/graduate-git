## smoothing

```python
def smoothing(data0,order=4):
    data=data0.copy()
    for i in range(114):
        y=data.iloc[i,2:].values.astype("int")
        x=np.arange(2000,2018)
        poly=np.polyfit(x,y,order)
        poly_y=np.poly1d(poly)(x)
        data.iloc[i,2:]=poly_y
    return data.iloc[:,2:]

ls_sm=smoothing(ls)
gdp_sm=smoothing(gdp)
pop_sm=smoothing(pop)
```

```python
def hann(data0,N=5):
    data=data0.copy()
    weights=np.hanning(N)
    print(weights)
    for i in range(114):
        c=data.iloc[i,2:].values
        sam=np.convolve(weights/weights.sum(),c)[N-1:-N+1]
        data.iloc[i,4:4+14]=sam
    return data.iloc[:,4:4+14]

ls_ha=hann(ls)
gdp_ha=hann(gdp)
pop_ha=hann(pop)
lspop_ha=hann(lspop)

smooth_ha={0:ls_ha,1:gdp_ha,2:pop_ha}
smooth_hals={0:ls_ha,1:gdp_ha,2:lspop_ha}
```

```python
final_models=[]
ts=13
for cluster in range(4):
    cluster=0
    r2=np.empty(ts)
    coef1=np.empty(ts)
    coef2=np.empty(ts)
    inter=np.empty(ts)
    ts_dic=np.arange(ts)+1
    bestr2=[-100,0]
    bestmodel=None
    
    for time_step in range(1,ts+1):
#     for time_step in [9]:
        df={}
        for para in range(3):
            datas=smooth_ha[para]
            data1=datas.iloc[lb==cluster,:-time_step].values.reshape(-1)
            data2=datas.iloc[lb==cluster,time_step:].values.reshape(-1)
            dif=(data2-data1)/data1
            df.update({para:dif})

        dataset=pd.DataFrame(df)
        
        model = LinearRegression()
        train_X=dataset.iloc[:,[1,2]]
        train_Y=dataset.iloc[:,[0]]
        model.fit(train_X, train_Y)
        train_pred_Y = model.predict(train_X)
        rmse_train = np.sqrt(mean_squared_error(train_pred_Y, train_Y))

        r2score=r2_score(train_Y,train_pred_Y)
        # msle_train = mean_squared_log_error(train_pred_Y, train_Y)
        
        
        n=train_Y.size
        p=2
        adj_r2score=1-(1-r2score)*(n-1)/(n-1-p)
        r2[time_step-1]=adj_r2score
        coef1[time_step-1]=model.coef_[0][0]
        coef2[time_step-1]=model.coef_[0][1]
        inter[time_step-1]=model.intercept_[0]
        if bestr2[0]<r2score:
            bestr2=[r2score,time_step]
            bestmodel=model
            bestdata=dataset
#             plt.figure(figsize=(4,4))
#             axes = plt.gca()
#             lim=[-0.0, 1.5]
#             axes.set_xlim(lim)
#             axes.set_ylim(lim)
#             plt.plot(train_pred_Y,train_Y,'o')
#             plt.plot(lim,lim)
    print(bestr2)
    final_models.append(bestmodel)
    models=pd.DataFrame({"time_step":ts_dic,"adj_r2":r2,"coef_gdp":coef1,"coef_pop":coef2,"intercept":inter})
    break
print(n,p)
models
```
