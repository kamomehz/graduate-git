```python
import numpy as np
import matplotlib.pyplot as plt
import glob, os
import pandas as pd
```

```python
sl_ua=pd.read_csv("sl16_50.csv").iloc[:,1:]
ls_ua=pd.read_csv("lsfinal1999.csv").iloc[:,8:]
ls_ua
```

```python
for ct in [0,2,3,4,19]:
    ssp=np.ones((5,9))
    years=list(range(2010,2051,5))
    name=sl_ua.iloc[ct,0]
    
    for i in range(9):
        datax=pd.read_csv("predictCSV/%d.csv"%i).iloc[:,1:]
        datax=sl_ua.iloc[:,[0]].merge(datax,on="city")
        ssp[:,i]=datax.iloc[ct,1:6].values

    plt.figure(figsize=(10,7))
    plt.title("%s"%name)
    
    for i in range(5):
        plt.plot(years,ssp[i],"o-",label="SSP %d"%(i+1))
        
    sl_year=list(range(2016,2051))
    sl_x=sl_ua.iloc[ct,1:].values
    plt.plot(sl_year,sl_x,"-",label="SLEUTH")
    
    ls_year=list(range(2016,2051))
    ls_x=ls_ua.iloc[ct,:].values
    plt.plot(sl_year,sl_x,"-",label="SLEUTH")
    
    plt.legend()
    plt.savefig("resultFig/%d_%s2.png"%(ct,name))
#     break
```
