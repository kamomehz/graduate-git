```python
g2015=np.ones(184)
p2015=np.ones(193)
g2010=np.ones(184)
p2010=np.ones(193)
regionG=list(range(184))
regionP=list(range(193))
for i in range(184):
    g2015[i]=sspGDP.iloc[i*5:i*5+5,6].sum()/5
    g2010[i]=sspGDP.iloc[i*5,5]
    regionG[i]=sspGDP.iloc[i*5,2]
for i in range(193):
    p2015[i]=sspPOP.iloc[i*5:i*5+5,6].sum()/5
    p2010[i]=sspPOP.iloc[i*5,5]
    regionP[i]=sspPOP.iloc[i*5,2]
```

```python
datag2015=pd.DataFrame({'Region':regionG,'g2010':g2010,'g2015': g2015})
datap2015=pd.DataFrame({'Region':regionP,'p2010':p2010,'p2015': p2015})
```

```python
histGDP2=histGDP.copy()
histGDP2=pd.merge(histGDP2,datag2015, on='Region')
histPOP2=histPOP.copy()
histPOP2=pd.merge(histPOP2,datap2015, on='Region')
```

```python
def int2str(x,option):
    l=x.shape[0]
    out=['Region']
    for i in range(l):
        out.append(option+str(x[i]))
    return np.array(out)
```

```python
sspGDP.columns
```
